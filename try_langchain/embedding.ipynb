```python
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
text = [
    "Langchain is a powerful library for building applications with large language models.",
    "Embeddings are numerical representations of text that capture semantic meaning.",
    "FAISS is a library for efficient similarity search and clustering of dense vectors.",
    "HuggingFace provides a wide range of pre-trained models for natural language processing tasks"
]
```

```python
embedded_modal = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_texts(text, embedded_modal)
vector_store.save_local("my_first_vector_db")
```

```python
retriever = vector_store.as_retriever()
query = "What is Langchain?"
docs = retriever.get_relevant_documents(query)
print(docs)
```

```python
for doc in docs:
    print(doc.page_content)
```
